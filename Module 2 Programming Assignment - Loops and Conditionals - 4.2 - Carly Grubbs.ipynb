```python
## This program will display a list of produce and state true/false if they are small
## It will also print true/false if the fruits are large

print("We have a list of produce: pea, cherry, watermelon, and pumpkin.")
print("We will let you know if the produce is small and green, small and not green, not small and green, or not small and not green.")

print()  ## Printing blank space

# Starting with green and small
print("This produce is small and green")

## Setting variables
small = True
green = True

## Creating if..elif..else statements to compare the size and color and printing results
if small and green:
    print('pea')
elif small and not green:
    print('cherry')
elif not small and green:
    print('watermelon')
elif not small and not green:
    print('pumpkin')

print()  ## Printing blank space

## Small but not green
print("This produce is small but not green")

## Setting variables
small = True
green = False

## Creating if..elif..else statements to compare the size and color and printing results
if small and green:
    print('pea')
elif small and not green:
    print('cherry')
elif not small and green:
    print('watermelon')
elif not small and not green:
    print('pumpkin')

print()  ## Printing blank space

## Not small but green
print("This produce is not small but green")

## Setting variables
small = False
green = True

## Creating if..elif..else statements to compare the size and color and printing results
if small and green:
    print('pea')
elif small and not green:
    print('cherry')
elif not small and green:
    print('watermelon')
elif not small and not green:
    print('pumpkin')

print()  ## Printing blank space

## Not small and not green
print("This produce is not small and not green")

## Setting variables
small = False
green = False

## Creating if..elif..else statements to compare the size and color and printing results
if small and green:
    print('pea')
elif small and not green:
    print('cherry')
elif not small and green:
    print('watermelon')
elif not small and not green:
    print('pumpkin')
```
